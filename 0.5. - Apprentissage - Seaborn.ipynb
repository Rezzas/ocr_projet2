# Introduction

Seaborn est une librairie qui vient s'ajouter à Matplotlib, remplace certains réglages par défaut et fonctions, et lui ajoute de nouvelles fonctionnalités. Seaborn vient corriger trois défauts de Matplotlib:

* Matplotlib, surtout dans les versions avant la 2.0, ne génère pas des graphiques d'une grande qualité esthétique.
* Matplotlib ne possède pas de fonctions permettant de créer facilement des analyses statistiques sophistiquées.
* Les fonctions de Matplotlib ne sont pas faites pour interagir avec les Dataframes de Pandas.

Seaborn fournit une interface qui permet de palier ces problèmes. Il utilise toujours Matplotlib "sous le capot", mais le fait en exposant des fonctions plus intuitives.

* Les graphes les plus utilisés :
  * pairplot
  * catplot
  * boxplot
  * distplot
  * jointplot
  * heatmap

* Structure quasiment commune à toutes les fonctions Seaborn :

## Iris Dataset

```python
import seaborn as sns

df = sns.load_dataset('iris')
df.head()
```

```python
sns.pairplot(df, hue='species')
```

* Immédiatement l'on remarque que pour l'espèce "setosa" il y a une relation forte entre la taille des sépales (ce qui supporte la pétale) et la taille des pétales.

* Les histogrammes donnent la distribution que suit telle ou telle variable. On voit bien que l'espèce Setosa n'est en lien avec aucune autre espèce pour les pétales.
  * sepal_width suit une distribution normale 


```python
sns.boxplot(x='sepal_width',y='sepal_length', data=df, hue='species')
```

```python
sns.distplot(df['sepal_width'], rug=True)
```

* La largeur des sépales suit une distribution normale.
* Dit autrement : il y a de fortes chances pour que la plupart des sépales soient comprises entre 1 écart type de différence par rapport à la moyenne

```python
sns.jointplot(x='sepal_width',y='sepal_length', data=df)
```
